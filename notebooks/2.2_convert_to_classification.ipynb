```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/compteur-one-hot-encoded.csv", index_col=0)
df.head()
```

```python
classes = pd.cut(df['Comptage horaire'], bins=(-1, 2, 7, 15, 26, 39, 54, 74, 107, 170, 250, df['Comptage horaire'].max()), labels=["0-2", "3-7", "8-15", "16-26", "27-39", "40-54", "55-74", "75-107", "108-170", "171-250", "251+"])
classes.value_counts(sort=False, normalize=True)
```

```python
df = df.drop(columns=['Comptage horaire'])
df['Comptage horaire'] = classes
```

```python
df.head()
```

```python
df.to_csv('../data/processed/compteur-classes-one-hot-encoded.csv')
```

```python
df = pd.read_csv("../data/processed/lieu-compteur-one-hot-encoded.csv", index_col=0)
df.head()
```

```python
classes = pd.cut(df['Comptage horaire'], bins=(-1, 3, 9, 18, 31, 46, 64, 86, 115, 155, 230, 450, df['Comptage horaire'].max()), 
                 labels=["0-3", "4-9", "10-18", "19-31", "32-46", "47-64", "65-86", "87-115", "116-155", "156-230", "231-450", "451+"])
vc = classes.value_counts(sort=False)
vc
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=vc)
plt.xticks(rotation=45)
plt.xlabel("Classes de comptage horaire")
plt.ylabel("Nombre de valeurs")
plt.show()
```

```python
df = df.drop(columns=['Comptage horaire'])
df['Comptage horaire'] = classes
```

```python
df.head()
```

```python
df.to_csv('../data/processed/lieu-compteur-classes-one-hot-encoded.csv')
```
